#### Ejercicio 18. Normaliza la variable de precios para cada población y traza los 4 histogramas en el mismo gráfico (debes usar el subconjunto obtenido en la pregunta 14) (★★★)

Para el método de normalización, puedes usar el que consideres adecuado, no hay una única respuesta correcta para esta pregunta. Imprime el gráfico y escribe en la celda de Markdown un breve análisis sobre el gráfico.

Pista: Puedes ayudarte revisando la demostración multihist de Matplotlib.

```python
indice_fuenla = cs[cs['level5'] == 'Fuenlabrada'].index
cs.loc[indice_fuenla, 'price_normalized'] = (cs.loc[indice_fuenla, 'price'] - cs.loc[indice_fuenla, 'price'].min()) / (cs.loc[indice_fuenla, 'price'].max() - cs.loc[indice_fuenla, 'price'].min())
precios_normalizados_fuenla = cs.loc[indice_fuenla, 'price_normalized'].dropna().astype(float)


indice_lega = cs[cs['level5'] == 'Leganés'].index
cs.loc[indice_lega, 'price_normalized'] = (cs.loc[indice_lega, 'price'] - cs.loc[indice_lega, 'price'].min()) / (cs.loc[indice_lega, 'price'].max() - cs.loc[indice_lega, 'price'].min())
precios_normalizados_lega = cs.loc[indice_lega, 'price_normalized'].dropna().astype(float)

indice_geta = cs[cs['level5'] == 'Getafe'].index
cs.loc[indice_geta, 'price_normalized'] = (cs.loc[indice_geta, 'price'] - cs.loc[indice_geta, 'price'].min()) / (cs.loc[indice_geta, 'price'].max() - cs.loc[indice_geta, 'price'].min())
precios_normalizados_geta = cs.loc[indice_geta, 'price_normalized'].dropna().astype(float)

indice_alco = cs[cs['level5'] == 'Alcorcón'].index
cs.loc[indice_alco, 'price_normalized'] = (cs.loc[indice_alco, 'price'] - cs.loc[indice_alco, 'price'].min()) / (cs.loc[indice_alco, 'price'].max() - cs.loc[indice_alco, 'price'].min())
precios_normalizados_alco = cs.loc[indice_alco, 'price_normalized'].dropna().astype(float)


plt.figure(figsize = (10, 5))
plt.hist(precios_normalizados_fuenla, bins = 30, alpha = 0.7, label='Fuenlabrada')
plt.hist(precios_normalizados_lega, bins = 30, alpha = 0.7, label='Leganés')
plt.hist(precios_normalizados_geta, bins = 30, alpha = 0.7, label='Getafe')
plt.hist(precios_normalizados_alco, bins = 30, alpha = 0.7, label='Alcorcón')
plt.title("Histograma de precios normalizados del Cinturón Sur de Madrid")
plt.legend()
plt.show()
```

En el histograma actual de precios normalizados del cinturón sur de Madrid, podemos observar como las cuatro localidades concentran la mayor parte de las viviendas en precios que se mantienen entre el 0 y el 0.4 de la gráfica, dandonos a entender que se concentran más en precios bajos que altos. Siendo Leganés la localiad con menos concentración en el lado derecho y con una distribución en el gráfico más repartida, y dando a entender que sus precios, aunque tambien se concentran en mayor parte en los valores bajos, tambien varian hacia precios mas altos con más frecuencia que las demás localidades.

#### Ejercicio 21. Realiza un trazado de las coordenadas (columnas latitud y longitud) del cinturón sur de Madrid por color de cada población (debes usar el subconjunto obtenido del Ejercicio 14) (★★★★)

Ejecuta la siguiente celda y luego comienza a codear en la siguiente. Debes implementar un código simple que transforme las columnas de coordenadas en un diccionario de Python (agrega más información si es necesario) y agrégala al mapa.

```python
from ipyleaflet import Marker, Icon, basemaps, Map, AwesomeIcon

map = Map(center = (40.4168, -3.7038), zoom = 10, min_zoom = 1, max_zoom = 20, 
    basemap=basemaps.OpenStreetMap.Mapnik)
```

```python
grupo = cs.groupby('level5')[['latitude', 'longitude']].first() # Agrupar coordenadas por localidad.
diccionario = grupo.to_dict(orient='index')# Convertirlas en un diccionario.

for key, datos in diccionario.items():    # Como las cordenadas aparecen en el diccionario separadas por ',' y no por '.' se las toma como un string
                                         # y no como float y no las acepta el mapa, por eso las convertimos a float. 
    datos['latitude'] = float(datos['latitude'].replace(',', '.')) 
    datos['longitude'] = float(datos['longitude'].replace(',', '.'))

for key, datos in diccionario.items(): # Las agregamos al mapa. 
    marcador = Marker(location=(datos["latitude"], datos["longitude"]), title=key)
    map.add_layer(marcador)

map
```
